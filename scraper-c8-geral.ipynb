## **Scraper para obter os dados do novo mapeamento**

### **49er e 49erFX**

- World Championship 2020: https://49er.org/events/2020-world-championship/#result-49
- World Championship 2019: https://49er.org/events/2019-world-championship/#result-49
- European Championship 2021: https://49er.org/events/2021-european-championship/#result-49

```python
PATH_TO_CHROMEDRIVER = 'chromedriver-mac-x64/chromedriver'

options = Options()
options.add_argument("--headless")

driver = Chrome(PATH_TO_CHROMEDRIVER)
```

```python
df_49er_WC_2020, df_49erFX_WC_2020 = cluster8(
    URL = 'https://49er.org/events/2020-world-championship/#result-49',
    PATH_TO_CHROMEDRIVER = PATH_TO_CHROMEDRIVER,
    NOME_COMPETICAO = 'World Championship 2020',
    ID_COMPETICAO = 14,
    DRIVER = driver
)
```
