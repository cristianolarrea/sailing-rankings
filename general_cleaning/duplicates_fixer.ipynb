```python
import pandas as pd
import numpy as np

df = pd.read_excel('ilca7-wc2015.xlsx')
```

```python
len(df)
```

```python
# get unique "Nome Competidor"
names = df['Nome Competidor'].unique()

# create dict with names as keys and number of entries as values
names_dict = {}
for name in names:
    names_dict[name] = len(df[df['Nome Competidor'] == name])

# divide all values by 2

for key in names_dict:
    names_dict[key] = names_dict[key] / 2
```

```python
# create df2 where for each name (key) we keep the first #values entries
df2 = pd.DataFrame()

for key in names_dict:
    df2 = df2.append(df[df['Nome Competidor'] == key].head(int(names_dict[key])))

len(df2)
```

```python
# output df2 to excel
df2.to_excel('ilca7-wc2015-clean.xlsx', index=False)
```
